```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/nn_train.csv')
data.head()
```

```python
x = data.iloc[:,1:1025]
print(x.head())
x = (x - x.mean())/x.std()
print(x.head())
x = x.values
print(x.shape)
y = data.iloc[:,1025]
nan_count = y.isna().sum()
print(f"Number of NaN values: {nan_count}")
y = y.values
print(y.shape)
print(type(y))
```

```python
m = x.shape[0]
n = int(0.2*m)
x_test = x[:n,:].T
print(x_test.shape)
x = x[n:,:].T
print(x.shape)
y_test = y[:n].T
print(y_test.shape)
y_train = y[n:].T
print(y_train.shape[0])
```

```python
def cost_function(a3,y,l,w1,w2,w3):

    e = 1e-10  # for safety
    m = y.shape[0]
    cost  = (1/m)*(np.sum(-y*np.log(a3 + e)-(1-y)*(np.log(1-a3 + e)))) + (l/(2*m))*(np.sum(np.square(w1))+ np.sum(np.square(w2)) + np.sum(np.square(w3)))
    return cost
```

```python

n_x = x.shape[0]
n_h1 = 1200
n_h2 = 1000
n_y = 1
alpha = 0.001
iteration = 150
l = 0.0001
parameters = []
w1 = np.random.randn(n_h1,n_x)*np.sqrt(2/n_x)
parameters.append(w1)
b1 = np.zeros((n_h1, 1))
parameters.append(b1)
w2 = np.random.randn(n_h2,n_h1)*np.sqrt(2/n_h1)
parameters.append(w2)
b2 = np.zeros((n_h2,1))
parameters.append(b2)
w3 = np.random.randn(n_y,n_h2)*0.01
parameters.append(w3)
b3 = np.zeros((n_y,1))
parameters.append(b3)
def relu(x):
    return np.maximum(x, 0)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def d_relu(x):
    return np.array(x > 0, dtype = np.float32)
```

```python


def prediction(x):

    z1 = np.dot(w1,x) + b1
    a1 = relu(z1)
    z2 = np.dot(w2,a1) + b2

    a2 = relu(z2)
    z3 = np.dot(w3,a2) + b3
    a3 = sigmoid(z3)
    print(a3.shape)
    a3_bool = (a3 > 0.5).astype(int)
    return a3_bool
# Initialize Adam parameters
beta1 = 0.9  # Exponential decay for the first moment estimates
beta2 = 0.999  # Exponential decay for the second moment estimates
epsilon = 1e-8  # Small constant to avoid division by zero
alpha_initial = 0.001
l = 0.001 # Learning rate

# Initialize moment estimates for all parameters
m_w1, m_w2, m_w3 = 0, 0, 0  # First moments for weights
v_w1, v_w2, v_w3 = 0, 0, 0  # Second moments for weights
m_b1, m_b2, m_b3 = 0, 0, 0  # First moments for biases
v_b1, v_b2, v_b3 = 0, 0, 0  # Second moments for biases
decay_rate = 0.96
decay_step = 100
t = 0  # Time step for bias correction

j_history = []
for i in range(iteration):
    # Forward propagation
    z1 = np.dot(w1, x) + b1
    a1 = relu(z1)
    z2 = np.dot(w2, a1) + b2
    a2 = relu(z2)
    z3 = np.dot(w3, a2) + b3
    a3 = sigmoid(z3)

    # Backward propagation
    m = x.shape[1]
    dz3 = (a3 - y_train)
    dw3 = (1/m) * (np.dot(dz3, a2.T) + l * np.sum(w3))
    db3 = (1/m) * np.sum(dz3, axis=1, keepdims=True)
    dz2 = (1/m) * np.dot(w3.T, dz3) * d_relu(a2)
    dw2 = (1/m) * (np.dot(dz2, a1.T) + l * np.sum(w2))
    db2 = (1/m) * np.sum(dz2, axis=1, keepdims=True)
    dz1 = (1/m) * np.dot(w2.T, dz2) * d_relu(a1)
    dw1 = (1/m) * (np.dot(dz1, x.T) + l * np.sum(w1))
    db1 = (1/m) * np.sum(dz1, axis=1, keepdims=True)

    # Update Adam parameters
    t += 1
    alpha = (alpha_initial)*(decay_rate)**(t//decay_step)
    # Update weights and biases using Adam
    for param, grad, m, v in zip(
        [w1, w2, w3, b1, b2, b3],
        [dw1, dw2, dw3, db1, db2, db3],
        [m_w1, m_w2, m_w3, m_b1, m_b2, m_b3],
        [v_w1, v_w2, v_w3, v_b1, v_b2, v_b3],
    ):
        m = beta1 * m + (1 - beta1) * grad  # Update biased first moment
        v = beta2 * v + (1 - beta2) * (grad ** 2)  # Update biased second moment
        m_corrected = m / (1 - beta1 ** t)  # Correct bias for first moment
        v_corrected = v / (1 - beta2 ** t)  # Correct bias for second moment
        param -= alpha * m_corrected / (np.sqrt(v_corrected) + epsilon)  # Update parameter

    # Compute and store cost
    if i < iteration:
        c = cost_function(a3, y_train, l, w1, w2, w3)
        j_history.append(c)

    # Print cost at intervals
    if (i % (iteration // 10)) == 0:
      y_pred = prediction(x)
      accuracy = np.mean(y_train == y_pred)
      print(f"Iteration {i}, Cost: {c}",accuracy : { accuracy * 100})

# Plot cost history
plt.plot(j_history)
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost Reduction Over Iterations")
plt.show()
```
